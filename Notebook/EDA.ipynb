### 2.2. Dataset Information
- **Gender:** This attribute represents the gender of the individual and can have two possible values: Female or Male.
- **Age:** This attribute is numeric and represents the age of the individual. It is a continuous variable.
- **Height:** This attribute is numeric and represents the height of the individual in some unit of measurement (e.g., centimeters or inches).
- **Weight:** This attribute is numeric and represents the weight of the individual in some unit of measurement (e.g., kilograms or pounds).
- **family_history_with_overweight:** This attribute indicates whether the individual has a family history of overweight. It can have two possible values: yes or no.
- **FAVC:** FAVC stands for "Frequent consumption of high-caloric food." It represents whether the individual frequently consumes high-caloric food and can have two possible values: yes or no.
- **FCVC:** This attribute is numeric and represents the frequency of consumption of vegetables. It is a continuous variable.
- **NCP:** NCP stands for "Number of main meals." It represents the number of main meals the individual consumes in a day. It is a numeric attribute.
- **CAEC:** CAEC stands for "Consumption of food between meals." It represents the frequency of consuming food between meals and can have four possible values: no, Sometimes, Frequently, or Always.
- **SMOKE:** This attribute represents whether the individual smokes or not and can have two possible values: yes or no.
- **CH2O:** CH2O stands for "Consumption of water daily." It represents the amount of water the individual consumes daily. It is a numeric attribute.
- **SCC:** SCC stands for "Calories consumption monitoring." It represents whether the individual monitors their calorie consumption or not and can have two possible values: yes or no.
- **FAF:** FAF stands for "Physical activity frequency." It represents the frequency of physical activity the individual engages in and is a numeric attribute.
- **TUE:** TUE stands for "Time using technology devices." It represents the amount of time the individual spends using technology devices (e.g., smartphones, computers) and is a numeric attribute.
- **CALC:** CALC stands for "Consumption of alcohol." It represents the frequency of alcohol consumption and can have four possible values: no, Sometimes, Frequently, or Always.
- **MTRANS:** MTRANS stands for "Mode of transportation." It represents the mode of transportation used by the individual and can have five possible values: Automobile, Motorbike, Bike, Public_Transportation, or Walking.
- **NObeyesdad:** This attribute represents the obesity level of the individual and can have seven possible values: Insufficient_Weight, Normal_Weight, Overweight_Level_I, Overweight_Level_II, Obesity_Type_I, Obesity_Type_II, or Obesity_Type_III

## 4. Exploring Data (Visualization)
- Normal - Normal_Weight
- Underweight - Insufficent_Weight
- Overweight - Overweight_Level_I, Overweight_Level_II
- Obesity - Obsesity_Type_I, Obsesity_Type_II
- Extreme Obesity - Obsesity_Type_III

```python
mapping = {
    'Normal': ['Normal_Weight'],
    'Underweight': ['Insufficient_Weight'],
    'Overweight': ['Overweight_Level_I', 'Overweight_Level_II'],
    'Obesity': ['Obesity_Type_I', 'Obesity_Type_II'],
    'Extreme Obesity': ['Obesity_Type_III']
}

reverse_mapping = {value: key for key, values in mapping.items() for value in values}
df['NObeyesdad'] = df['NObeyesdad'].replace(reverse_mapping)
```

```python
df.NObeyesdad.unique()
```

### 4.1. Histogram and KDE

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df,x='Weight',bins=60,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='Weight',kde=True,hue='Gender')
plt.show()
```

#### Insight:
- Weight of the females appears to be greater than males in this dataset

```python
plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='Weight',kde=True,hue='NObeyesdad')
plt.subplot(142)
sns.histplot(data=df[df.Gender=='Female'],x='Weight',kde=True,hue='NObeyesdad')
plt.subplot(143)
sns.histplot(data=df[df.Gender=='Male'],x='Weight',kde=True,hue='NObeyesdad')
plt.show()
```

#### Insights:
- In this Datset Males tends to be more overweight and obese than females but Extreme obesity can be observed more in    females
- From this dataset it can be observed generally males tends to more fit than females
- From this datset it can be clearly observed the chance being underweight is more for females than male
